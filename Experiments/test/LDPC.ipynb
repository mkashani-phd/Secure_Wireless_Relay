```python
import channelCoding
import numpy as np
len_msg = 256
codeword_length = 2048

msg = np.random.randint(0, 2, len_msg)
codeword = channelCoding.encode_LDPC(msg, Codeword_length=codeword_length)




llr = np.array([10 if x == 0 else -10 for x in codeword])

decoded_msg = channelCoding.decode_LDPC(llr, len_msg).tolist()



print(list(decoded_msg) == list(msg))
```
