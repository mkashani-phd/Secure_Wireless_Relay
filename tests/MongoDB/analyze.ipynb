```python
import pymongo


# Connect to the MongoDB, change the connection string per your MongoDB environment
conn = pymongo.MongoClient("mongodb://10.29.162.146:27017")
db = conn['MAC_SC']
collection_dest_phase1 = db['destination, phase_1']
collection_dest_phase2 = db['destination, phase_2']
collection_relay = db['relay, phase_1']

conf = conn['config']['config'].find_one()
```

```python
total_dest_1 = collection_dest_phase1.count_documents(filter={})
Correctly_received = collection_dest_phase1.count_documents(filter={'message': conf['PAYLOAD']})

print(f"Total dest pahse 1: {total_dest_1}, Correctly received: {Correctly_received}")
```

```python
total_realy = collection_relay.count_documents(filter={})
Correctly_received = collection_relay.count_documents(filter={'$expr': {'$eq': ['$message', '$config.PAYLOAD']}})

print(f"Total relay: {total_realy}, Correctly received: {Correctly_received}")
```

```python
total_dest_2 = collection_relay.count_documents(filter={})
Correctly_received = collection_relay.count_documents(filter={'$expr': {'$eq': ['$message', '$config.PAYLOAD']}})
dest_phase_2_itegrity_verified = collection_dest_phase2.count_documents(filter={'integrity':True})

print(f"Total relay: {total_realy}, Correctly received: {Correctly_received}, dest phase 2 integrity verified: {dest_phase_2_itegrity_verified}")
```
